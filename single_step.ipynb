```python
import os
import numpy as np
from PIL import Image
```

```python
base_dir = "exp/ImageNet/batch_results/realistic_p2p_imagenet-r_sd15_caption_no_prompt/"
# base_dir = "exp/ImageNet/batch_results/cartoon_p2p_coco_sd15_caption_no_prompt/"
# base_dir = "exp/ImageNet/batch_results/cartoon_p2p_imagenet_sd15_caption_no_prompt/"
img_dirs = os.listdir(base_dir)
```

```python
for dir_ in img_dirs:
    img_dir = os.path.join(base_dir,dir_)
    images = os.listdir(img_dir)
    for image in images:
        image_path = os.path.join(img_dir, image)
        if "pnp" in image_path and "clip" not in image_path:
            img_np = np.array(Image.open(image_path))
            h, w, c = img_np.shape
            img = Image.fromarray(img_np[:,-512:,:])
            image_path = image_path[:-4]+"_clip"+image_path[-4:]
            img.save(image_path)
```
